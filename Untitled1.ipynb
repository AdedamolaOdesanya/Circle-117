```python
def drop_redundant_columns(data):
    """
    Drop redundant columns from teh Chicago Crime Dataset.

    Parameters:
        df (pandas.DataFrame): Input DataFrame.

    Returns:
        pandas.DataFrame: DataFrame with redundant columns dropped.
    """
    # Create a set to store the column names
    seen = set()
    # Create a list to store columns to drop
    to_drop = []

    # Iterate through each column in the DataFrame
    for col in data.columns:
        # If the column is already in the set, it's redundant and should be dropped
        if col in seen:
            to_drop.append(col)
        else:
            seen.add(col)

    # Drop the redundant columns
    df = data.drop(columns=to_drop)

    return df
```

```python
from datetime import datetime

def extract_and_map_month(date):
    # ensure the input is in datetime format
    date_column = pd.to_datetime(date, format='%m/%d/%Y %I:%M:%S %p')
    
    month_names = date_column.dt.month_name()
    
    return month_names

def extract_and_map_day(date):
    # ensure the input is in datetime format
    date_column = pd.to_datetime(date, format='%m/%d/%Y %I:%M:%S %p')
    
    day_names = date_column.dt.day_name()
    
    return day_names
```

```python
def create_season_column(date):
    date_column = pd.to_datetime(date, format='%m/%d/%Y %I:%M:%S %p', errors='coerce')
    
    month_to_season = {
        1: 'Winter', 2: 'Winter', 3: 'Spring',
        4: 'Spring', 5: 'Spring', 6: 'Summer',
        7: 'Summer', 8: 'Summer', 9: 'Fall',
        10: 'Fall', 11: 'Fall', 12: 'Winter'
    }
    
    seasons_column = date_column.dt.month.map(month_to_season)
    
    return seasons_column
```

```python
chicago_1['Day'] = extract_and_map_day(chicago_1['Date'])
chicago_1['Month'] = extract_and_map_month(chicago_1['Date'])
chicago_1['Season'] = create_season_column(chicago_1['Date'])
```
